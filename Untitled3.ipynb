```python
import numpy as np
data1=np.genfromtxt("mushrooms.csv",delimiter=",",dtype=None)
data=data1[1:]
d=data.T

for i in d:
    m=set(i)
    if len(m)==2:
        s1=(list(m)[0])
        s2=(list(m)[1])
        i[i==s1]=1
        i[i==s2]=0 
  
y=d[0].T
x=d[1::].T
# print(x)
def gini(x,y):

    s=np.array([x, y])
    dat=s.T
    zeroes=dat[dat[:,0]=="0"]
    ones=dat[dat[:,0]=="1"]

    zero_true=zeroes[zeroes[:,1]=="0"]
    zero_false=zeroes[zeroes[:,1]=="1"]
    
    one_true=ones[ones[:,1]=="0"]
    one_false=ones[ones[:,1]=="1"]

    if len(zero_true)+len(zero_false)==0:
         gini1 = 1 - 0
    else:
        gini1 = 1 - ((len(zero_true)/(len(zero_true)+len(zero_false)))**2)-((len(zero_false)/(len(zero_true)+len(zero_false)))**2)

    if len(one_true)+len(one_false)==0:
        gini2 = 1 - 0
    else:    
        gini2= 1 - ((len(one_true)/(len(one_true)+len(one_false)))**2)-((len(one_false)/(len(one_true)+len(one_false)))**2)
    weight=(gini1*(len(zeroes)/len(dat)))+(gini2*(len(ones)/len(dat)))
                                                               
    return weight
def split(x,y):
    names=[]
    values=[]
    m=x.T
    for i in range(len(m)):
        if len(list(set(m[i] )))==2:
            print(data1[0][i+1]," : ",gini(m[i],y))
            names.append(str(data1[0][i+1]))
            values.append(gini(m[i],y))

    s=min(values)
    ind=values.index(s)
    print("min --> ",names[ind]," : ",s)

    
split(x,y)

          
```
